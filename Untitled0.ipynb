```
import pandas as pd
import statsmodels.formula.api as sms
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```
# Lendo os dados
boston = pd.DataFrame.from_csv("https://raw.githubusercontent.com/selva86/datasets/master/BostonHousing.csv")
boston.head()
```

```
adv = boston
adv.head
```

```
model_medv = sms.ols(formula="lstat ~ medv", data=adv).fit()
print(model_medv.summary())
```

```
model_medv = sms.ols(formula="age ~ medv", data=adv).fit()
print(model_medv.summary())
```

```
model_medv = sms.ols(formula="lstat ~ medv + age", data=adv).fit()
print(model_medv.summary())
```
